##  Load packages

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
```

```python
%cd location/of/web/driver
url = f"https://bit.ly/3s4lcG3"
```

```python
def extract_review_details(review_elem):
  # author name
  try:
    author = review_elem.find_element_by_class_name("bv-author").get_attribute('innerText')
  except NoSuchElementException:
    author = ""

  # star rating
  star = -1
  metas = review_elem.find_elements_by_tag_name("meta")
  for meta in metas: 
    if meta.get_attribute("itemprop") == "ratingValue":
      star = meta.get_attribute("content")
  
  # title
  try:
    title = review_elem.find_element_by_css_selector(".bv-content-title-container > h3").get_attribute('innerText')
  except NoSuchElementException:
    title = ""

  # body
  try:
    review_body = review_elem.find_element_by_css_selector(".bv-content-summary-body-text > p").get_attribute('innerText')
  except NoSuchElementException:
    review_body = ""
  
  # product recomendation
  try:
    recomend = review_elem.find_element_by_css_selector("li.bv-content-data-value > span.bv-content-data-label").get_attribute('innerText')
  except NoSuchElementException:
    recomend = ""

  return {'author': author, 'star': star, 'title': title, 'review_body': review_body, 'recomend': recomend}
```

```python
def next_reviews_page(driver):
  # extract review frame pagination text, if the text is changed that mean the page has been loaded 
  current_page = driver.find_element_by_css_selector("#bv-content-pagination-pages-reviewContentList1-7 > span").get_attribute('innerText')
  driver.find_element_by_css_selector("li.bv-content-pagination-buttons-item-next > a").click()
  # check if the text changed
  WebDriverWait(driver, delay).until(lambda x : current_page != driver.find_element_by_css_selector("#bv-content-pagination-pages-reviewContentList1-7 > span").get_attribute('innerText'))
  
```

```python
driver = webdriver.Chrome(executable_path='chromedriver.exe')
driver.get(url)
delay = 15 #seconds
# Close manually all popups
```

```python
reviews_list=[]
while True:
  # wait until review frame appear
  myElem = WebDriverWait(driver, delay).until(EC.visibility_of_element_located((By.CSS_SELECTOR, ".bv-content-list-reviews")))

  # extract only reviews frame to fetch content fast
  reviews = driver.find_elements_by_class_name("bv-content-top-review")
  for review in reviews:
    reviews_list.append(extract_review_details(review))

  # check is there a next button 
  if len(driver.find_elements_by_css_selector("li.bv-content-pagination-buttons-item-next > a")) > 0:
    next_reviews_page(driver)
  else:
    break
  
```

```python
pd_reviews = pd.DataFrame(reviews_list)
pd_reviews
```
